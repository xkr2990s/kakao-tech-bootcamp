### 과제 2: 분류 분석
목표: 특정 데이터에 대한 나이브 베이즈 분류 모델을 구축하고 평가합니다.

데이터셋: Wine dataset을 사용합니다.

내용:
1. 데이터를 불러오고, 독립 변수(X)와 종속 변수(y)를 설정합니다.
2. 훈련 데이터와 테스트 데이터로 분할합니다.
3. 나이브 베이즈 모델을 학습시키고, 테스트 데이터에 대해 예측합니다.
4. 예측 결과를 평가하고, 정확도와 혼동 행렬을 계산합니다.
5. 혼동 행렬을 시각화합니다.


```python
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# 데이터 로드
wine = load_wine()
X = wine.data
y = wine.target
print(f'x: {X}')
print(f'y: {y}')

# 데이터 분할
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

# 모델 학습
nb = GaussianNB()
nb.fit(X_train, y_train)

# 예측
y_pred = nb.predict(X_test)
print(f'Predict: {y_pred}')

# 평가
accuracy = accuracy_score(y_test,y_pred)
conf_matrix = confusion_matrix(y_test,y_pred)
print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
```

```python
# 시각화
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

###과제 3: 교차 검증

목표: 데이터에 대해 교차 검증을 수행하여 모델 성능을 평가합니다.

데이터셋: Diabetes dataset을 사용합니다.

내용:
1. 데이터를 불러오고, 독립 변수(X)와 종속 변수(y)를 설정합니다.
2. 나이브 베이즈 모델을 사용하여 5-겹 교차 검증을 수행합니다.
3. 교차 검증 점수를 출력하고 평균 점수를 계산합니다.
4. 교차 검증 결과를 시각화합니다.


```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import cross_val_score
```

```python
# 데이터 로드
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target



# 모델 생성
nb = GaussianNB()

# 교차 검증
scores = cross_val_score(nb,X,y,cv=5)
print(f'Cross-validation scores: {scores}')
print(f'Mean CV Score: {np.mean(scores)}')
```

```python
# 시각화
plt.plot(range(1,len(scores)+1),scores,marker='o',linestyle='--')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Scores')
plt.show()
```
